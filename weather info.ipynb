```python
import glob
import os
```

```python
path = "./ghcnd_hcn/ghcnd_hcn/"
all_files = glob.glob(os.path.join(path, "*.dly"))

df_dict = dict()
for f in all_files:

    df = pd.read_fwf(f, widths=[11,4,2,4,5,1,1,1,5,1,1,1,5,1,1,1,5,1,1,1,5,1,1,1,5,1,1,1,5,1,1,1,5,1,1,1,5,1,1,1,5,1,
                           1,1,5,1,1,1,5,1,1,1,5,1,1,1,5,1,1,1,5,1,1,1,5,1,1,1,5,1,1,1,5,1,1,1,5,1,1,1,5,1,1,1,
                           5,1,1,1,5,1,1,1,5,1,1,1,5,1,1,1,5,1,1,1,5,1,1,1,5,1,1,1,5,1,1,1,5,1,1,1,5,1,1,1,5,1,1,1],
                   header=None)
    df = df[(df[1]>=2000)&(df[1]<=2005)]
    
    df = df.rename(columns={0:"ID",1:"YEAR",2:"MONTH",3:"ELEMENT"})
    
    day_of_month=1
    
    for i in range(4,128):
        if i%4 == 0:
            df=df.rename(columns={i:"VALUE"+str(day_of_month)})
        elif i%4 == 1:
            df=df.rename(columns={i:"MFLAG"+str(day_of_month)})
        elif i%4 == 2:
            df=df.rename(columns={i:"QFLAG"+str(day_of_month)})
        else:
            df=df.rename(columns={i:"SFLAG"+str(day_of_month)})
            day_of_month +=1
    
    df_dict[f] = df
```
